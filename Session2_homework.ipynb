### 5.Illustrate relationship between Total Bags and Small Bags using scatter plot.

```python
plt.xlabel("Total Bags"),
plt.ylabel('Small Bags')
plt.scatter(x= 'Total Bags', y='Small Bags', data=avacado_data, color = 'blue')
plt.legend()
```

### 6.Construct AveragePrice distribution.

### 7.Which type of Avocados are expensive?(with boxplot)

```python
sns.boxplot(x = 'AveragePrice', y = 'type', data = avacado_data, palette = 'rainbow')
```

### 8.Construct heatmap correlation matrix.

```python
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(avacado_data.corr(), annot = True, linewidths= 0.5, fmt ='.1f')
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.title('Heatmap of Avacado')
plt.savefig('Heatmap_of_Avacado.png')
plt.show()
```

### 9.Find relationship between large and small bags using scatterplot.(use jointplot function)

```python
sns.jointplot(x='Large Bags', y='Small Bags', data = avacado_data, kind = 'scatter')
```

### 10.Find linear relationship between Total Volume and Bags for each avocado type.(use lmplot function)

```python
sns.lmplot(x='Total Volume',y='Small Bags',data = avacado_data, col = 'type')
sns.lmplot(x = "Total Volume", y = "Large Bags", data = avacado_data, col = 'type' )
sns.lmplot(x = "Total Volume", y = "XLarge Bags", data = avacado_data, col = 'type')
```
