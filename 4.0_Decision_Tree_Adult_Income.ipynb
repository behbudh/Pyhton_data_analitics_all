```python
# Create Dummy variables
X_income = dataset_Income.drop('IncomeClass', 1)
Y_income = dataset_Income.IncomeClass
X_income
```

```python
Y_income
```

```python
# Create X and Y Variables
X_income = pd.get_dummies(X_income)
Y_income = pd.get_dummies(Y_income)
X_income
```

```python
Y_income
```

```python
x = X_income
y = Y_income
```

```python
# Split the X and Y dataset into training and testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
# Import and train classifier
from sklearn import tree
dtm = tree.DecisionTreeClassifier(criterion = 'entropy')
dtm.fit(x_train, y_train)
```

```python
# Test the model
pred_dtm = dtm.predict(x_test)
pred_dtm
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import multilabel_confusion_matrix
acc_dtm = accuracy_score(y_test, pred_dtm)
matrix_dtm = multilabel_confusion_matrix(y_test, pred_dtm)
acc_dtm
```
